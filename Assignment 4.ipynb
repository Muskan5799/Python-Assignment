Question 1:
Write a Python program that checks if a number entered by the user is even or odd. Print "Even" if the number is even and "Odd" if it is odd.




```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

Question 2:
Create a program that prompts the user for their age and prints whether
they are a child (0-12), teenager (13-19), or adult (20 and above).

```python
age = int(input("Enter your age: "))
if age <= 12:
    print("Child")
elif age <= 19:
    print("Teenager")
else:
    print("Adult")
```

Question 3:
Write a program that takes two numbers as input from the user and prints which number is larger. If both numbers are equal, print "Both numbers are equal.

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
if a > b:
    print("First number is larger")
elif b > a:
    print("Second number is larger")
else:
    print("Both numbers are equal")
```

Question 5:
Write a program that takes a score from the user (0-100) and prints the corresponding grade:

A for 90-100

B for 80-89

C for 70-79

D for 60-69

F for below 60

```python
score = int(input("Enter your score (0-100): "))
if 90 <= score <= 100:
    print("Grade A")
elif 80 <= score <= 89:
    print("Grade B")
elif 70 <= score <= 79:
    print("Grade C")
elif 60 <= score <= 69:
    print("Grade D")
else:
    print("Grade F")
```


Question 8:
Create a program that checks if a user-entered character is a vowel (a, e, i, o, u) or a consonant. Print the corresponding message.

```python
ch = input("Enter a single character: ").lower()
if ch in ['a', 'e', 'i', 'o', 'u']:
    print("Vowel")
elif ch.isalpha():
    print("Consonant")
else:
    print("Not a valid alphabet character")
```

Question 9:
Write a program that checks if a given year is a leap year. A year is a leap year if it is divisible by 4, except for years that are divisible by 100 but not divisible by 400.

```python
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

Question 10:
Create a program that prompts the user for a temperature in Celsius and converts it to Fahrenheit. If the converted temperature is greater than 100Â°F, print "It's boiling!" Otherwise, print "It's not boiling yet.

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit:", fahrenheit)
if fahrenheit > 100:
    print("It's boiling!")
else:
    print("It's not boiling yet.")
```
