Question 2:
Assign the number 12.5 to a variable price . Print both the variable and its
data type. Then assign the value 12 to the same variable and print its data
type again. What do you observe?

```python
price = 12.5
print("Value of price:", price)
print("Data type of price:", type(price))

price = 12
print("Value of price after reassignment:", price)
print("Data type of price after reassignment:", type(price))
```

Question 3:

Question 6: Assign values to x and y, print their types, sum them, and print the result

```python
x = 10
y = 3.5
print("x:", x, type(x))
print("y:", y, type(y))
sum_xy = x + y
print("Sum of x and y:", sum_xy, type(sum_xy))
```
